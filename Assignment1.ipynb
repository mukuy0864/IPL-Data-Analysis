```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")

matches.head()
```

```python
#i) Match won by Maximum Margin of Runs

max_run_win = matches.loc[matches['win_by_runs'].idxmax()]
max_run_win[['season','team1','team2','winner','win_by_runs']]

plt.bar(max_run_win['winner'], max_run_win['win_by_runs'])
plt.title("Highest Run Margin Win")
plt.show()

# Insight:
# This match had the highest run difference in IPL history.
# Shows strong batting + strong bowling performance.
# Big wins create more audience excitement.
```

```python
#ii) Match won by Maximum Wickets
max_wicket_win = matches.loc[matches['win_by_wickets'].idxmax()]
max_wicket_win[['season','team1','team2','winner','win_by_wickets']]

# Insight:
# Winning by 10 wickets means complete domination.
# Shows powerful chasing team performance.
```

```python
#iii) Match won by minimum margin of runs

matches[matches['win_by_runs'] > 0].sort_values('win_by_runs').head(1)

# Insight:
# These matches were very close.
# Such matches increase audience excitement.
```

```python
#iii) Match won by Minimum Runs
min_run_win = matches[matches['win_by_runs'] > 0]
min_run_win=min_run_win[min_run_win['win_by_runs']==min_run_win['win_by_runs'].min()]
min_run_win[['season','team1','team2','winner','win_by_runs']]

# Insight:
# Very close match.
# Such matches increase fan engagement.
```

```python
#iv) Match won by Minimum Wickets
wicket_wins = matches[matches['win_by_wickets'] > 0]

# Get match with minimum wickets
min_wicket_win = wicket_wins[wicket_wins['win_by_wickets'] == wicket_wins['win_by_wickets'].min()]

# Show result
min_wicket_win[['winner', 'win_by_wickets']]

# Insight:
# Nail-biting finishes attract high viewership.
```

```python
#v) Matches where D/L Method Applied
matches['dl_applied'].value_counts()



# Insight:
# D/L method is used during rain-affected matches.
# Helps maintain fairness.
```

```python
#vi) Number of Matches in Each City
city_matches = matches['city'].value_counts()
city_matches.head()

city_matches.head(10).plot(kind='')
plt.title("Top 10 Cities by Matches Hosted")
plt.show()

# Insight:
# Some cities host more matches due to stadium size and fan base.
# Metro cities attract more IPL matches.
```

```python
#vii) Matches Won by Each Team
team_wins = matches['winner'].value_counts()
team_wins

team_wins.plot(kind='bar')
plt.title("Total Wins by Each Team")
plt.show()

# Insight:
# Teams with highest wins are most successful franchises.
# These teams have stronger fan following.
```

```python
#viii) Matches Played Every Season
season_matches = matches['season'].value_counts().sort_index()
season_matches

season_matches.plot(kind='line')
plt.title("Matches Played Per Season")
plt.show()

# Insight:
# Increase in matches shows league expansion.
# More teams = more matches = more revenue.
```

```python
#ix) Top 10 Players by Man of Match
top_mom = matches['player_of_match'].value_counts().head(10)
top_mom

top_mom.plot(kind='barh')
plt.title("Top 10 Players - Man of the Match")
plt.show()

# Insight:
# These players are consistent performers.
# Star players increase brand value of IPL.
```

```python
#x) Does Winning Toss Mean Winning Match?
toss_match_win = matches[matches['toss_winner'] == matches['winner']]
percentage = (len(toss_match_win)/len(matches))*100
print(percentage)

# Insight:
# If percentage is around 50-55%, toss has slight impact.
# Skill matters more than toss.
```

```python
#xi) Toss Decision (Bat or Field)
matches['toss_decision'].value_counts()

sns.countplot(x='toss_decision', data=matches)
plt.title("Toss Decision")
plt.show()

# Insight:
# Most captains prefer chasing (field first).
# Dew factor influences decision.
```

```python
#xii) D/L per Season
dl_season = matches.groupby('season')['dl_applied'].sum()
dl_season.plot(kind='line',marker='o')
plt.title("D/L Per Season")
plt.show()
```

```python
#xiii) Different Match Results
matches['result'].value_counts()
```

```python
#xiv) Times Each Team Won Toss
matches['toss_winner'].value_counts().plot(kind='barh')
plt.title("Times Each Team Won Toss")
plt.show()
```

```python
#xv & xvi) Best Venue & Best Teams for Chasing/Defending
#Defending Team:
defending = matches[matches['win_by_runs'] > 0]['winner'].value_counts()
#Chasing Team:
chasing = matches[matches['win_by_wickets'] > 0]['winner'].value_counts()
print(defending)
print(chasing)
# Insight:
# Some teams perform better while defending.
# Some teams specialize in chasing targets.
```

```python
#xvii) Matches Played in Each Stadium
venue_matches = matches['venue'].value_counts()
venue_matches.head(10)
```
